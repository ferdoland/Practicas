## Video1 . Introduccion
Una serie de tiempo es un conjunto de observaciones tomadas **Secuencialmente en el tiempo**
Los **datos tabulares** son aquellos que podemos desordenar las filas y la calidad de la información no varía , 
en las series de tiempo el orden de las filas es inalterable y si es alterada , la información también es dañada .

Las series de tiempo tienen tecnicas propias de análisis y predicción

Pricipales tecnicas de tratamiento:

1.- Analisis exploratorio
    graficación e identificación gráfica de patrones y sucesoso.
2.- Pronósticos o "Forecasting".
    Creación de modelos predictivos de comportamientos futuros.
